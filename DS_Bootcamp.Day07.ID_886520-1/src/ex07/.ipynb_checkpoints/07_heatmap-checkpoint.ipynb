```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable

conn = sqlite3.connect('../data/checking-logs.sqlite')
```

```python
weekday_query = """
SELECT uid, 
       strftime('%w', timestamp) AS weekday, 
       COUNT(*) AS commits
FROM checker
WHERE uid LIKE 'user_%'
GROUP BY uid, weekday
"""
weekday_df = pd.io.sql.read_sql(weekday_query, conn)
weekday_pivot = weekday_df.pivot(index='weekday', columns='uid', values='commits').fillna(0)

weekday_pivot = weekday_pivot[weekday_pivot.sum().sort_values(ascending=False).index]

weekday_names = ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']
weekday_pivot.index = weekday_names

hour_query = """
SELECT uid, 
       strftime('%H', timestamp) AS hour, 
       COUNT(*) AS commits
FROM checker
WHERE uid LIKE 'user_%'
GROUP BY uid, hour
"""
hour_df = pd.io.sql.read_sql(hour_query, conn)
hour_pivot = hour_df.pivot(index='hour', columns='uid', values='commits').fillna(0)

hour_pivot = hour_pivot[hour_pivot.sum().sort_values(ascending=False).index]
```

```python
plt.figure(figsize=(15, 5))
ax = plt.gca()
im = ax.imshow(weekday_pivot, cmap='coolwarm')
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
plt.colorbar(im, cax=cax)
ax.set_xticks(range(len(weekday_pivot.columns)))
ax.set_xticklabels(weekday_pivot.columns, rotation=90)
ax.set_yticks(range(len(weekday_pivot.index)))
ax.set_yticklabels(weekday_pivot.index)
ax.set_title('Commits per User by Weekday')
plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 8))
ax = plt.gca()
im = ax.imshow(hour_pivot, cmap='coolwarm')
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
plt.colorbar(im, cax=cax)
ax.set_xticks(range(len(hour_pivot.columns)))
ax.set_xticklabels(hour_pivot.columns, rotation=90)
ax.set_yticks(range(len(hour_pivot.index)))
ax.set_yticklabels(hour_pivot.index)
ax.set_title('Commits per User by Hour')
plt.tight_layout()
plt.show()
```

* Which user has the most commits on Tue? The answer: user_2.
* Which user has the most commits on Thu? The answer: user_4.
* On which weekday do the users not like making a lot of commits? The answer: Fri.
* Which user at which hour made the largest number of commits? The answer: user_4, 21
