```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('../data/ab-test.csv')
before_newsfeed = df[df['time'] == 'before']
after_newsfeed = df[df['time'] == 'after']
control_group = df[df['group'] == 'control']
```

```python
plt.figure(figsize=(15, 8))
plt.title('Lag before labs deadlines. After and before the Newsfeed.', loc='center', y=1.05, fontsize=15)
plt.grid(False)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.boxplot([before_newsfeed[before_newsfeed['group'] == 'control']['diff'],
    before_newsfeed[before_newsfeed['group'] == 'test']['diff']], showmeans=True, patch_artist=True,
    boxprops=dict(facecolor='white', color='green', linewidth=2, alpha=0.8),
    medianprops=dict(color='red', linewidth=1),
    meanprops=dict(markeredgecolor='green', markerfacecolor='green'), 
    whiskerprops=dict(color='red', linewidth=1), 
    capprops=dict(color='red', linewidth=1), 
    labels=['control', 'test'])
plt.xlabel("[group]")
plt.grid(True)
plt.title('before')

plt.subplot(1, 2, 1)
plt.boxplot([after_newsfeed[after_newsfeed['group'] == 'control']['diff'],
    after_newsfeed[after_newsfeed['group'] == 'test']['diff']], showmeans=True, patch_artist=True,
    boxprops=dict(facecolor='white', color='green', linewidth=2, alpha=0.8),
    medianprops=dict(color='red', linewidth=1),
    meanprops=dict(markeredgecolor='green', markerfacecolor='green'),
    whiskerprops=dict(color='red', linewidth=1), 
    capprops=dict(color='red', linewidth=1), 
    labels=['control', 'test'])
plt.xlabel("[group]")
plt.grid(True)
plt.title('after')

plt.tight_layout()
plt.show()
```

What was the IQR of the control group before the newsfeed?

```python
Q1_before = before_newsfeed[before_newsfeed['group'] == 'control']['diff'].quantile(0.25)
Q3_before = before_newsfeed[before_newsfeed['group'] == 'control']['diff'].quantile(0.75)
IQR_before = Q3_before - Q1_before
print(IQR_before)
```
