```python
import pandas as pd
import sqlite3

conn = sqlite3.connect('../data/checking-logs.sqlite')
```

```python
query = """
SELECT timestamp
FROM checker
WHERE uid LIKE "user_%"
"""

df=pd.io.sql.read_sql(query, conn)
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])

df['date'] = df['timestamp'].dt.date
df['hour'] = df['timestamp'].dt.hour
df['day_of_week'] = df['timestamp'].dt.day_name()

working_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
weekends = ['Saturday', 'Sunday']

df['day_type'] = df['day_of_week'].apply(lambda x: 'Working Day' if x in working_days else 'Weekend')
```

```python
result = pd.DataFrame()
weeknd_hourly_counts = df[df['day_type']== 'Weekend'].groupby(['date', 'hour']).size().reset_index(name='count')
result['weekend'] = weeknd_hourly_counts.groupby('hour')['count'].mean()
```

```python
daily_hourly_counts = df[df['day_type']== 'Working Day'].groupby(['date', 'hour']).size().reset_index(name='count')
result['working_day'] = daily_hourly_counts.groupby('hour')['count'].mean()
```

```python
result.plot(kind='bar', subplots=True, 
             title='Commits per hour', fontsize=8, figsize=(15,8), color=['blue', 'green'])
```
