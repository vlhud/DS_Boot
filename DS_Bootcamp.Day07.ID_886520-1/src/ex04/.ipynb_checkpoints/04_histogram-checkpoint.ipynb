```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

conn = sqlite3.connect('../data/checking-logs.sqlite')
```

```python
query = """
SELECT timestamp
FROM checker
WHERE uid LIKE "user_%"
"""

df=pd.io.sql.read_sql(query, conn)
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])

df['hour'] = df['timestamp'].dt.hour
df['day_of_week'] = df['timestamp'].dt.day_name()

working_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
weekends = ['Saturday', 'Sunday']

working_day_commits = df[df['day_of_week'].isin(working_days)].groupby('hour').size()
weekend_commits = df[df['day_of_week'].isin(weekends)].groupby('hour').size()
```

```python
plt.figure(figsize=(15,8))

plt.hist(working_day_commits.index, weights=working_day_commits.values, bins=24, alpha=0.7, label='working days', color='blue')
plt.hist(weekend_commits.index, weights=weekend_commits.values, bins=24, alpha=0.7, label='weekends', color='orange')
plt.legend()
plt.show()
```

Are there hours when the total number of commits was higher on weekends than on working days?

```python
comparison = pd.DataFrame({
    'Working Day Commits': working_day_commits,
    'Weekend Commits': weekend_commits
})

higher_weekend_commits = comparison[comparison['Weekend Commits'] > comparison['Working Day Commits']]

top_4_hours = higher_weekend_commits.nlargest(4, 'Weekend Commits')
print(top_4_hours)
```
