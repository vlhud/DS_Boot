```python
import pandas as pd
import sqlite3

conn = sqlite3.connect('../data/checking-logs.sqlite')
```

```python
query_views = """
SELECT 
    DATE(datetime) AS date
FROM pageviews
WHERE uid like 'user_%' 
"""
df_views = pd.read_sql_query(query_views, conn)

query_commits = """
SELECT 
    DATE(timestamp) AS date
FROM checker
WHERE uid like 'user_%' 
"""
df_commits = pd.read_sql_query(query_commits, conn)
```

```python
daily_views = df_views.groupby('date').size().reset_index(name='views')

daily_commits = df_commits.groupby('date').size().reset_index(name='commits')

merged_data = pd.merge(daily_views, daily_commits, on='date', how='inner')
merged_data.set_index('date', inplace=True)

pl=merged_data.plot(title="Views and commits per day", fontsize=8, figsize=(15,8), rot=90, kind='line', style=['ro-.', 'c--'],)
pl.grid(axis='x', linestyle='-', linewidth='0.5', color='gray')
```

How many times was the number of views larger than 150?

```python
views_above_150 = (merged_data['views'] > 150).sum()
print(f"The answer is {views_above_150}")
```
