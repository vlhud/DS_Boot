```python
import pandas as pd

df = pd.read_json('../auto.json')
df.set_index('CarNumber', inplace=True)

# Строки, где штрафы превышают 2100
fines_above_2100 = df[df['Fines'] > 2100]
print("Строки, где штрафы превышают 2100:")
print(fines_above_2100)

# Отобразить строки, в которых сумма штрафов превышает 2100, а сумма возврата равна 2
fines_above_2100_refund_2 = df[(df['Fines'] > 2100) & (df['Refund'] == 2)]
print("\nСтроки, в которых сумма штрафов превышает 2100, а сумма возврата равна 2:")
print(fines_above_2100_refund_2)

# Отобразить строки, в которых указаны модели из списка:  ['Focus', 'Corolla']
models_list = ['Focus', 'Corolla']
models_selected = df[df['Model'].isin(models_list)]
print("\nСтроки, в которых находятся модели из списка ['Focus', 'Corolla']:")
print(models_selected)

#  Строки, в которых указан номер автомобиля из указанного списка
car_numbers_list = ['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']
car_numbers_selected = df[df.index.isin(car_numbers_list)]
print("\nСтроки, в которых указан номер автомобиля из указанного списка:")
print(car_numbers_selected)

# Отображение средних значений штрафов, сгруппированных по марке
median_fines_by_make = df.groupby('Make')['Fines'].median()
print("\nСредние штрафы, сгруппированные по марке:")
print(median_fines_by_make)

# Отображение средних значений штрафов, сгруппированных по марке и модели
median_fines_by_make_model = df.groupby(['Make', 'Model'])['Fines'].median()
print("\nСредние значения штрафов, сгруппированных по марке и модели:")
print(median_fines_by_make_model)

# Отображение количества штрафов, сгруппированных по марке и модели
count_fines_by_make_model = df.groupby(['Make', 'Model'])['Fines'].count()
print("\nКоличество штрафов, сгруппированных по марке и модели:")
print(count_fines_by_make_model)

# Отображение минимальных и максимальных штрафов, сгруппированных по марке и модели
min_max_fines_by_make_model = df.groupby(['Make', 'Model'])['Fines'].agg(['min', 'max'])
print("\nМинимальные и максимальные штрафы, сгруппированные по марке и модели:")
print(min_max_fines_by_make_model)

# Отображение стандартного отклонения значения штрафов, сгруппированных по марке и модели
std_fines_by_make_model = df.groupby(['Make', 'Model'])['Fines'].std()
print("\nСтандартное отклонение значения штрафов, сгруппированных по марке и модели:")
print(std_fines_by_make_model)

# Отображение номеров автомобилей, сгруппированных по количеству штрафов в порядке убывания
fines_count_by_car_number = df.groupby(df.index)['Fines'].count().sort_values(ascending=False)
print("\nНомера автомобилей сгруппированы по количеству штрафов в порядке убывания:")
print(fines_count_by_car_number)

# Выберите из исходного фрейма данных все строки, соответствующие номеру автомобиля, занявшего первое место
top_car_number = fines_count_by_car_number.index[0]
top_car_number_data = df.loc[top_car_number]
print(f"\nСтроки, соответствующие номеру автомобиля top-1 ({top_car_number}):")
print(top_car_number_data)

# Отображение номеров автомобилей, сгруппированных по сумме штрафов в порядке убывания
fines_sum_by_car_number = df.groupby(df.index)['Fines'].sum().sort_values(ascending=False)
print("\nНомера автомобилей сгруппированы по сумме штрафов в порядке убывания:")
print(fines_sum_by_car_number)

# Выберите из исходного фрейма данных все строки, соответствующие номеру автомобиля, занявшего первое место
top_car_number_sum = fines_sum_by_car_number.index[0]
top_car_number_sum_data = df.loc[top_car_number_sum]
print(f"\nСтроки, соответствующие номеру автомобиля, занявшего первое место по сумме штрафов ({top_car_number_sum}):")
print(top_car_number_sum_data)

# Выведите на экран таблицу, которая отвечает на вопрос: есть ли какие-либо автомобильные номера, которые были привязаны к разным моделям?
car_numbers_with_different_models = df.groupby(df.index)['Model'].nunique()
different_models = car_numbers_with_different_models[car_numbers_with_different_models > 1]
print("\nАвтомобильные номера, привязанные к разным моделям:")
print(different_models)
```
