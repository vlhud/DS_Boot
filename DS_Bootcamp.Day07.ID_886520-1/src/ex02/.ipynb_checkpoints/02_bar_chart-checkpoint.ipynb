```python
import pandas as pd
import sqlite3

conn = sqlite3.connect('../data/checking-logs.sqlite')
```

```python
query = """
SELECT timestamp
FROM checker
WHERE uid LIKE "user_%"
"""

df=pd.io.sql.read_sql(query, conn)

df['timestamp'] = pd.to_datetime(df['timestamp'])

def categorize_time(timestamp):
    if timestamp.hour < 4:
        return 'night'
    elif timestamp.hour < 10:
        return 'morning'
    elif timestamp.hour < 17:
        return 'afternoon'
    else:
        return 'evening'

df['time_of_day'] = df['timestamp'].apply(categorize_time)

df['date'] = df['timestamp'].dt.date
time_of_day_counts = df.groupby(['date', 'time_of_day']).size().unstack(fill_value=0)

palette=['blue','orange', 'green', 'brown']
pl=time_of_day_counts.plot(title="Commits per day", fontsize=8, figsize=(15,8), rot=90, kind="bar", stacked=True, color=palette)
pl.grid(axis='y', linestyle='-', linewidth='0.5', color='gray')
```

When do our users usually commit the labs: in the night, morning, afternoon,
or evening?

```python
most_common_periods = df['time_of_day'].value_counts().nlargest(2)
print(most_common_periods)
```

Which day has:

the most number of commits
and at the same time, the number of commits in the evening is higher than in the afternoon?

```python
time_of_day_counts = df.groupby(['date', 'time_of_day']).size().unstack(fill_value=0)
evening_afternoon_counts = time_of_day_counts[['evening', 'afternoon']]
most_commits_day = df.groupby('date').size()

valid_days = evening_afternoon_counts[evening_afternoon_counts['evening'] > evening_afternoon_counts['afternoon']]

most_commits_valid_day = most_commits_day[most_commits_day.index.isin(valid_days.index)].idxmax()
print(most_commits_valid_day)
```
