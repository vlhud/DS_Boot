```python
import pandas as pd
from pandas.plotting import scatter_matrix
import sqlite3

conn = sqlite3.connect('../data/checking-logs.sqlite')
```

```python
query = """
SELECT 
    AVG((strftime('%s', test.first_commit_ts) - deadlines.deadlines) / 3600) AS avg_diff,
    COUNT(DISTINCT pageviews.datetime) as pageviews,
    COUNT(DISTINCT checker.timestamp) as commits
FROM 
    test
JOIN 
    pageviews ON test.uid = pageviews.uid  
JOIN 
    deadlines ON test.labname = deadlines.labs
JOIN
    (SELECT * FROM checker WHERE labname != 'project1') as checker ON test.uid = checker.uid
WHERE
    test.labname != 'project1'
GROUP BY
    test.uid
"""
df = pd.io.sql.read_sql(query, conn)
```

```python
scatter_matrix(df, figsize=[15, 8], diagonal='kde', color='red', s=200, alpha=1, density_kwds={'lw': 3})
```

Can we say that if a user has a low number of pageviews then they likely have a low number of commits?

no

Can we say that if a user has a low number of pageviews then they likely have a small average difference between the first commit and the lab deadline?

no

Can we say that there are many users with a low number of commits and a few with a high number of commits

yes

Can we say that there are many users with a small average difference and a few with a large average difference
