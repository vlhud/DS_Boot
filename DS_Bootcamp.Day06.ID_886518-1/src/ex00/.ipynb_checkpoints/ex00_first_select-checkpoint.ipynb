```python
import pandas as pd
import sqlite3
```

## создайте подключение к базе данных с помощью библиотеки sqlite3

```python
conn = sqlite3.connect('../data/checking-logs.sqlite')
```

## получите схему таблицы просмотров страниц, используя pd.io.sql.read_sql и запрос "PRAGMA table_info(просмотры страниц)".

```python
schema_query = "PRAGMA table_info(pageviews);"
schema = pd.io.sql.read_sql(schema_query, conn)
print(schema)
```

## получите только первые 10 строк таблицы просмотров страниц, чтобы проверить, как выглядит таблица

```python
first_10_rows_query = "SELECT * FROM pageviews LIMIT 10;"
first_10_rows = pd.io.sql.read_sql(first_10_rows_query, conn)
print(first_10_rows)
```

## получите подтаблицу, используя только один запрос, где:

* используются только uid и datetime
* используются только пользовательские данные (user_*), а не данные администратора
* он отсортирован по uid в порядке возрастания
* индексный столбец — это дата и время
* datetime преобразуется в DatetimeIndex
* имя фрейма данных — просмотры страниц

```python
filtered_query = """
SELECT uid, datetime 
FROM pageviews 
WHERE uid like 'user_%' 
ORDER BY uid ASC;
"""
pageviews = pd.io.sql.read_sql(filtered_query, conn)

pageviews['datetime'] = pd.to_datetime(pageviews['datetime'])
pageviews.set_index('datetime', inplace=True)
print(pageviews)
```
